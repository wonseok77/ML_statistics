```python
# 실행 오류 시 : pip install statsmodels
import statsmodels.api as sm
```

```python
### 데이터 생성하기
### 데이터 생성하기
# 변속기 : am
# 마력 : hp
# 연비 : mpg
mtcars = sm.datasets.get_rdataset('mtcars').data
mtcars
```

```python
### 모델 라이브러리 생성하기
import statsmodels.formula.api as smf
```

```python
### 종속변수와 독립변수 정의하기
formula = 'am ~ mpg + hp'
# 모델 객체 생성하기
result = smf.logit(formula=formula, data=mtcars)

# 학습모델 생성하기
result = result.fit()
# Current fuction value : 독립변수가 분류 결정에 미치는 영향의 정도
# 0.5에 가까울 수록 좋음
# - 현재 모델의 성능점수 : 0.3 이라고 보시면 됩니다
# - literations : 결과를 도출하는데 반복한 횟수 (9번 반복해서 최종 학습 종료 ...)
# Optimization terminated successfully : 최적화 성공 이라는 메세지 ..
print('result.fit() 결과 = \n' , result)

### 학습 결과 확인하기
print('result 학습결과 ' , result.summary())
```

```python
# 머신러닝에서 사용되는 독립변수들은 독립적인 관계를 가져야 합니다.
# 연관성이 없을 수록 좋습니다.
# 단, 상관관계(인과성)는 높을 수록 좋습니다. (다른 데이터에 영향을 미치는 정도)
# P 값이 작아질수록 상관관계가 높아진다 ( 종속 변수와의 관계 )
### coef : 기울기
#  - 기울기 값이 (음수) 이면 : (자동 0을 선택할 확률이 높아짐)
#  - 기울기 값이 (양수) 이면 : (수동 1을 선택할 확률이 높아짐)
### stderr : 표준편차
### P>|Z| : p-value (유의미값) < 0.05 이면 좋음, 0.05보다 크면 안좋음
###  - p-value < 0.05 : '유의미하다' 라고 해석합니다
```

```python
### 테스트(예측) 하기

import numpy as np

# 테스트 데이터로 데이터프레임 전체 넣기...
pred = result.predict(mtcars)

print('예측값 = \n', np.around(pred))
print('----------------------------')
print('실제값 = \n', np.around(mtcars['am']))
```

```python
### ROC(Receiver Operating Characteristic : 모델 성능 평가 지표)
# - 여러 임계값들을 기준으로 재현율(Recall), 양선율(Fallout)의 변화를 시각화 한 것
# <성능평가 순서>
# 1. 혼동 행렬표(confustion_matrix) 생성 : 예측 결과 데이터 이용
# 2. 정확도, 정밀도, 재현율(TPR), 특이도, 양선율 (FPR) 계산
# 3. ROC Curve 시각화 : x축(양선율(FPR)), y축(재현율(TPR)) 사용
# 4. ROC 면적 계산
```

```python
### 분류 정확도 확인하기
conf_tab = result.pred_table()
print(conf_tab)


'''
                                    예측값
                     Positive(긍정)        negative(부정)
실제값  참(True)       [[16(정탐),                    3(미탐)]
        거짓(False)    [3(오탐),             10(일상적 탐지)]]
'''
'''
적군(1긍정) 아군(0부정)일 경우

-TP(True Positive)
    : 실제값 적군을 예측값 적군으로 잘 예측한 개수(정확하게 탐지)
    
-TN(True Negative)
    : 실제값 적군을 예측값 아군으로 잘못 예측한 개수(미탐지 - 적이 뚫고 들어옴)
    : 재현율에 영향을 미침(조금 나쁨, 2종오류라고 칭함)
    
-FN(False negative)
    : 실제값 아군을 예측값 아군으로 잘 예측한 개수(일상적인 탐지)
    
-FP(False Positive)
    : 실제값 아군을 예측값 적군으로 잘못 예측한 개수(오탐지 - 아군끼리 싸움)
    : 정밀도 및 특이도에 영향을 미치는 값()
```

```python
### 분류 정확도 계산하기
# 공식 : (TP + TN) / 전체 개수 = 정확도(accuracy)
accuracy = (conf_tab[0][0] + conf_tab[1][1]) / len(mtcars)
print('분류 정확도 = ', accuracy)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np

accuracy_score = accuracy_score(mtcars['am'], np.around(pred))
print('라이브러리 정확도 = ' , accuracy_score)
```

```python
## 정밀도
# - 공식 : TP / (TP + FP)
# - 예측을 얼마나 정밀하게 하는지
# - 정밀도는 0~1 사이의 값을 가지며, 높을 수록 좋음
# - 정밀도가 1이라면, FP(오탐)과 0건 이라는 의미(오탐이 없다)
precision = conf_tab[0][0] / (conf_tab[0][0] + conf_tab[1][0])
print('정밀도 = ', precision)


### 재현율(recall : TPR)
# - 공식 : TP / (TP + FN)
# - 양성인 케이스에 대해 양성으로 잘 예측한 비율
# - 값이 1에 가까울 수록 좋음
# - 재현율이 1이면, FN이 0건이라는 의미(미탐이 없다는 의미)
recall_TPR = conf_tab[0][0] / (conf_tab[0][0] + conf_tab[0][1])
print('재현율 = ' , recall_TPR)


### 양선율(fallout: FPR)
# - 공식 : FP / FP + TN
# - 음성인 케이스에 대해 양성으로 잘못 예측한 비율
#   (정상을 비정상으로 판단한 경우)
fallout_FPR = conf_tab[1][0] / (conf_tab[1][0] + conf_tab[0][1])
```
